```python
import pandas as pd
import numpy as np
```

```python
# Task 1 -Create Project DataFrame

project = pd.DataFrame({
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    "Project": ["Project 1","Project 2","Project 3","Project 4","Project 5","Project 6","Project 7","Project 8","Project 9","Project 10","Project 11","Project 12","Project 13","Project 14"],
    "Cost": [1002000,2000000,4500000,5500000,np.nan,680000,400000,350000,np.nan,300000,2000000,1000000,3000000,200000],
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished","Failed","Finished","Failed","Ongoing","Finished","Failed","Ongoing","Finished","Finished"]
})

project
```

```python
# Task 2 - Create Employee DataFrame

employee = pd.DataFrame({
    "ID": ["A001","A002","A003","A004","A005"],
    "Name": ["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],
    "Gender": ["M","F","M","F","F"],
    "City": ["Paris","London","Berlin","Newyork","Madrid"],
    "Age": [25,27,29,31,30]
})

employee
```

```python
# Task - 3 Create Seniority Level DataFrame

seniority = pd.DataFrame({
    "ID": ["A001","A002","A003","A004","A005"],
    "Designation Level": [2,2,3,2,3]
})

seniority
```

```python
# Save DataFrames as CSV files

project.to_csv("project.csv", index=False)
employee.to_csv("employee.csv", index=False)
seniority.to_csv("seniority.csv", index=False)

print("All CSV files saved successfully!")
```

```python
# Fill missing Cost values using running average

cost_list = project["Cost"].tolist()

for i in range(len(cost_list)):
    if pd.isna(cost_list[i]):
        cost_list[i] = sum(cost_list[:i]) / len(cost_list[:i])

project["Cost"] = cost_list

project
```

```python
# Split Name into First Name and Last Name

employee[["First Name", "Last Name"]] = employee["Name"].str.split(" ", n=1, expand=True)

employee.drop("Name", axis=1, inplace=True)

employee
```

```python
# Task - 4 Join all three DataFrames

Final = pd.merge(project, employee, on="ID")
Final = pd.merge(Final, seniority, on="ID")

Final
```

```python
# Task 5 - Add Bonus Column

Final["Bonus"] = 0

Final.loc[Final["Status"] == "Finished", "Bonus"] = Final["Cost"] * 0.05

Final
```

```python
# Task 6 - Demote Failed Employees and Remove level Above 4

Final.loc[Final["Status"] == "Failed", "Designation Level"] = Final.loc[
    Final["Status"] == "Failed", "Designation Level"
] - 1

Final = Final[Final["Designation Level"] <= 4]

Final
```

```python
# Task 7 - Add Mr./Mrs. and drop gender coloumn

Final.loc[Final["Gender"] == "M", "First Name"] = "Mr. " + Final["First Name"]
Final.loc[Final["Gender"] == "F", "First Name"] = "Mrs. " + Final["First Name"]

Final.drop("Gender", axis=1, inplace=True)

Final
```

```python
# Task 8 - promote employees above age 29

for i in range(len(Final)):
    if Final.loc[i, "Age"] > 29:
        Final.loc[i, "Designation Level"] = Final.loc[i, "Designation Level"] + 1

Final
```

```python
# Task 9 - Create TotalProjCost DataFrame

TotalProjCost = Final.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

TotalProjCost.rename(columns={"Cost": "Total Cost"}, inplace=True)

TotalProjCost
```
